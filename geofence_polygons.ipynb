```python
import pandas as pd
import numpy as np
```

```python
site = pd.read_csv('site_prelevement_polygons.csv')
```

```python
site.head()
```

```python
site.rename(columns={"nom":"site_name"}, inplace=True)
```

```python
site.columns
```

```python
site["WKT"]  = site["WKT"].str.replace('POLYGON \(\(', '')
```

```python
site.loc[0, 'WKT']
```

```python
site["WKT"]  = site["WKT"].str.replace('\)\)', '')
```

```python
site.loc[0, 'WKT']
```

```python
site
```

```python
vertex = site['WKT'].str.split(',', expand=True)
```

```python
colonnes = [str(int(c) + 1) for c in list(vertex.columns)]
```

```python
vertex.columns = colonnes
```

```python
vertex = vertex.add_prefix('WKT.')
```

```python
vertex
```

```python
site = site.drop("WKT", axis=1)
```

```python
site
```

```python
polygons_df = pd.concat([site, vertex],axis = 1)
```

```python
polygons_df
```

```python
polygons_df_unpivot = pd.melt(polygons_df, id_vars=['id', 'site_name'])
```

```python
polygons_df_unpivot
```

```python
polygons_df_unpivot = polygons_df_unpivot.dropna()
```

```python
polygons_df_unpivot[['p_longitude', 'p_latitude']] = polygons_df_unpivot['value'].str.split(' ', expand=True)
```

```python
polygons_df_unpivot.head()
```

```python
polygons_df_unpivot = polygons_df_unpivot.sort_values(by=['id', 'site_name'])
```

```python
polygons_df_unpivot = polygons_df_unpivot.drop("value", axis=1)
```

```python
polygons_df_unpivot.head()
```

```python
polygons_df_unpivot.head()
```

```python
polygons_df_unpivot[['attribute', 'n']] = polygons_df_unpivot['variable'].str.split('.', expand=True)
```

```python
polygons_df_unpivot.head()
```

```python
polygons_df_unpivot = polygons_df_unpivot.drop(columns=["variable", "attribute"], axis=1)
```

```python
polygons_df_unpivot.head()
```

```python
polygons_df_unpivot['n'] = polygons_df_unpivot['n'].astype(int)
```

```python
polygons_df_unpivot.groupby('id')['n'].max()
```

```python
conditions = [
  (polygons_df_unpivot.id == 1),
  (polygons_df_unpivot.id == 2),
  (polygons_df_unpivot.id == 3),
  (polygons_df_unpivot.id == 4)
]
values = [11, 12, 10, 14]
polygons_df_unpivot["n_corner"] = np.select(conditions, values, default=0)
polygons_df_unpivot
```

```python
polygons_df_unpivot['id_key'] = polygons_df_unpivot["id"].apply(str) + '_' + polygons_df_unpivot['n'].apply(str)
```

```python
polygons_df_unpivot.head()
```

```python
polygons_df_unpivot = polygons_df_unpivot.drop(columns=['id', 'n'], axis=1)
```

```python
polygons_df_unpivot.head()
```

```python
polygons_df_unpivot.to_excel('site_prelevement_polygons_process.xlsx', index=False)
```
